# FEATURE EXTRACTION PROJECT

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('anime.csv')
df.head()
```

```python
def extract_episodes(title):
    check = False
    data = ""
    for i in title:
        if i == ")":
            break
        if check == True:
            data += i
        if i == "(":
            check = True
        
    return data
```

```python
df['Episodes'] = df['Title'].apply(extract_episodes)
df.head()
```

```python
df['Episodes'] = df['Episodes'].str.replace(' eps', '')
print(df['Episodes'].dtype)
```

```python
df['Episodes'] = df['Episodes'].astype(int)
print(df['Episodes'].dtype)
```

```python
def extract_time(title):
    check = False
    data = ""
    for i in range(len(title)):
        if title[i] == ')':
            for j in range(i+1, i+20):
                data += title[j]

            break
    return data
```

```python
df['Total_Time'] = df['Title'].apply(extract_time)
df.head()
```

```python
from datetime import datetime

def calculate_months(date_range):
    # split the string into start and end dates
    start_str, end_str = date_range.split(' - ')
    
    # parse the dates
    start = datetime.strptime(start_str.strip(), '%b %Y')
    end = datetime.strptime(end_str.strip(), '%b %Y')
    
    # calculate total months
    months = (end.year - start.year) * 12 + (end.month - start.month)
    
    return months

# apply to the column
df['Months'] = df['Total_Time'].apply(calculate_months)
```
