```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}
df = pd.DataFrame(data)
df
```

```python
data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]
column_label = ["Name", "Age", "City", "Salary"] 
list_df = pd.DataFrame(data_list, columns=column_label)
list_df
```

```python
# accessing single column from dataframe
list_df["Name"]
```

```python
# Accessing multiple column from dataframe
list_df[["Name", "Age","Salary"]]
```

```python
# Adding a new column to the dataframe.
list_df["Designation"] = ["Data Analyst", "Data Scientist", "AI Engineer", "RAG Specialist"]
list_df
```

```python
# Deleting a column from dataframe
# axis = 0 means column-wise
# axis = 1 means row-wise
list_df.drop('Designation', axis=1,inplace=True)
list_df
```
