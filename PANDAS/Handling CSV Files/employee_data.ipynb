Data Frame Methods:-

1. df.head() → Returns first 5 rows (or n rows) of the DataFrame

2. df.tail() → Returns last 5 rows (or n rows) of the DataFrame

3. df.sample() → Returns random rows from the DataFrame

4. df.info() → Gives summary of DataFrame (columns, data types, non-null counts)

5. df.shape → Returns tuple of (rows, columns)

6. df.describe() → Generates statistical summary (mean, std, min, max, etc.) for numeric columns

7. df.columns → Returns list of column names

8. df.nunique() → Returns count of unique values in each column
