```python
# HOW TO SELECT COLUMN?
df["city"]

# HOW TO SELECT MULTIPLE COLUMN
df[["city", "aqi"]]
```

loc --> selects on the basis of label
iloc --> selects on the basis of index

```python
# How to select ROWS?
df.loc[0]
df.loc[0:2]  #start_index(inclusive) : end_index(inclusive)

df.iloc[0]
df.iloc[0:2]  #start_index(inclusive) : end_index(exclusive)
```

```python
# How to select rows and columns?

df.loc[0:2, "city"]


# 3 Different methods to print same thing
df.loc[0:2, ["country", "city", "latitude", "longitude"]]
df.iloc[0:3, [1,2,3,4]]
df.iloc[0:3, 1:5]
```

```python
# How to select single scalar value? (at,iat)

df.at[0, 'city']
df.iat[0, 2]
```

FILTERING DATA

FILTERING DATA USING QUERY:-

query() makes filtering look like SQL and cleaner.

```python
# Expression as a string → Conditions are written inside quotes (e.g., "age > 25")
df.query('aqi > 200')

# Column based reference → Directly use column names without df['col'] (e.g., age > 25)

# Backticks (`) → Use when column names have spaces/special characters (e.g., `total marks` > 50)

# Operators → Supports &, |, ~, >, <, >=, <=, == for filtering conditions
df.query('temperature < 30 & aqi > 200')

# @ for Python variables → Use @ to access external variables inside query (e.g., age > @x)
Aqi = 200
df.query('aqi > @Aqi')

# Chained comparisons → Multiple conditions can be written cleanly (e.g., 20 < age < 30)
df.query('temperature < 30 & aqi > 200')[['city', 'country']]
```
